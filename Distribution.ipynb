```python
chance_tornado = 0.01
tornado_events = np.random.binomial(1, chance_tornado, 1000000)

two_days = 0
for i in range(1, len(tornado_events)-1):
    if tornado_events[i] == 1 and tornado_events[i-1] == 1:
        two_days += 1

print("{} tornados seguidos em {} anos".format(two_days, 1000000/365))
```
